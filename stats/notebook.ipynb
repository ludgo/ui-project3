```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    'tournament' : {
        'max' : 'tournament_max.txt',
        'median' : 'tournament_median.txt'
    },
    'roulette' : {
        'max' : 'roulette_max.txt',
        'median' : 'roulette_median.txt'
    }
}

for selection_method in data.keys():
    for attr, file in data[selection_method].items():
        values = pd.read_csv('./{}'.format(file))
    
        plt.plot(values)
        plt.xlabel(selection_method)
        plt.ylabel(attr)
        plt.show()
```
